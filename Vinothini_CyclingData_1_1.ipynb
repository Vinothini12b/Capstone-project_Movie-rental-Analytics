***`Task 1: In the above table add a column named “Day2(Km)” with the below Km numbers, and then print the table.

`***

```python
data["Day2(Km)"]=[9,3,8,8,5,10]
data
```

***`Task 2: Add a column named “Total” to the table. Total is calculated by the formula [Day1(Km)+Day2(Km)], and then print the table.

`***

```python
data["Total"]=data["Day1(Km)"]+data["Day2(Km)"]
data
```

***`Task 3: Add a column named “Day1Percent” to the table by calculating “Day1Percent”  and then print the table.

Where, “Day1Percent” = Day1(Km) / Total * 100

`***

```python
data["Day1Percent"]=data["Day1(Km)"] / data["Total"] * 100
data
```

***`Task 4: Add a column named “Day2Percent” to the table by calculating “Day2Percent”  and then print the table.

Where, “Day2Percent” = Day2(Km) / Total * 100

`***

```python
data["Day2Percent"]=data["Day2(Km)"] / data["Total"] * 100
data
```
