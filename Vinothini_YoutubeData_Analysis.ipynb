***`Task 1: Import the pandas library `***

```python
import pandas as pd
```

***`Task 2: Create a dataframe named “cwyd” from a given URL.

https://raw.githubusercontent.com/Invact-Abhay/DOE/main/doe2assignmentfinal.csv`***

```python
cwyd=pd.read_csv("https://raw.githubusercontent.com/Invact-Abhay/DOE/main/doe2assignmentfinal.csv")
cwyd
```

***`Task 3: Sort the data according to the ContentType in the alphabetical order.`***

```python
cwyd.sort_values("ContentType",ascending=True)
```

***`Task 4: Sort the data according to the Videotype in the descending alphabetical order.`***

```python
cwyd.sort_values("Videotype",ascending=False)
```

**`Task 5: Sort the data according to the Commentsadded in the ascending order`**

```python
cwyd.sort_values("Commentsadded",ascending=True)
```

***`Task 6: Sort the data according to the Shares in the descending order.`***

```python
cwyd.sort_values("Shares",ascending=False)
```

***`Task 7: Sort the data according to the Likes in the descending order and only show the Videotitle name.`***

```python
cwyd.sort_values("Likes",ascending=False)["Videotitle"]
```

***`Task 8: Sort the data based on the subscribers in ascending order and show only the viewers number of the sorted data.`***

```python
cwyd.sort_values("Subscribers",ascending=True)["Viewers"]
```

***`Task 9: Find the data of the video title where ContentType is Roast and sort the result on the basis of Videotitle name in alphabetical order.`***

```python
cwyd.query('ContentType=="Roast"').sort_values("Videotitle",ascending=True)
```

***`Task 10: Find the data of the video title where subscribers are more than 500 and sort the result on the basis of ContentType in alphabetical order. Show only video title names.`***

```python
cwyd.query('Subscribers> 500').sort_values("ContentType",ascending=True)["Videotitle"]
```

***`Task 11: Find the data of the videotitle with Impressions more than or equal to 4000 and then sort the data on the viewers value in descending order.`***

```python
cwyd.query('Impressions>=4000').sort_values("Viewers",ascending=False)
```

***`Task 12: Find the data of the product where Videotype is Long Video and then sort the Videotitle name in descending alphabetical  order. `***

```python
cwyd.query('Videotype=="Long Video"').sort_values("Videotitle",ascending=False)
```

***`Task 13: Find the name of the Videotitle whose ContentType is Stand-Up and then sort the data according to the Clicks in descending order. Get only the videotitle name for the same`***

```python
cwyd.query('ContentType=="Stand-Up"').sort_values("Clicks",ascending=False)["Videotitle"]
```

***`Task 14: Find the data where  Shares is more than the average Shares and then sort the data according to the impressions in descending order. Get only the videotitle name for the same.`***

```python
cwyd.query('Shares> Shares.mean()').sort_values("Impressions",ascending=False)["Videotitle"]
```

***`Task 15: Find the data where Subscribers are  less than the average subscribers and then sort the data according to the ContentType in alphabetical order. `***

```python
cwyd.query('Subscribers< Subscribers.mean()').sort_values("ContentType",ascending=True)
```
